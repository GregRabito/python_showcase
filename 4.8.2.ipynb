```python
import pandas as pd
import numpy as np
import os
```

```python
df_ords_prods_merge = pd.read_pickle(os.path.join(path, 'Data','Prepared Data', '4.8.1 orders_products_merge_final.pkl'))
```

# task

```python
df_ords_prods_merge.head(20)
```

```python
df_ords_prods_merge.groupby('department_id').agg({'order_number' : ['mean']})
```

```python
df_ords_prods_merge['customer_max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merge['customer_max_order'].head(40)
```

# flagging customer orders into loyalty categories

```python
df_ords_prods_merge.loc[df_ords_prods_merge['customer_max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['customer_max_order'] <= 40) & (df_ords_prods_merge['customer_max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['customer_max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df_ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(20)
```

```python
df_ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
df_ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean']})
```

```python
df_ords_prods_merge['customer_max_spent'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_ords_prods_merge['customer_max_spent'].head(20)
```

# flagging customer orders into spending categories

```python
df_ords_prods_merge.loc[df_ords_prods_merge['customer_max_spent'] >= 10, 'spending_amount'] = 'high_spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['customer_max_spent'] < 40, 'spending_amount'] = 'low_spender'
```

```python
df_ords_prods_merge[['user_id', 'spending_amount', 'prices']].head(20)
```

```python
df_ords_prods_merge['spending_amount'].value_counts(dropna = False)
```

```python
df_ords_prods_merge['customer_frequency'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df_ords_prods_merge['customer_frequency'].head(20)
```

```python
# flagging customer frequency into frequency categories
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['customer_frequency'] > 20, 'frequency_flag'] = 'non-frequent customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['customer_frequency'] <= 20) & (df_ords_prods_merge['customer_frequency'] > 10), 'frequency_flag'] = 'regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['customer_frequency'] <= 10, 'frequency_flag'] = 'frequent customer'
```

```python
df_ords_prods_merge[['user_id', 'customer_frequency', 'frequency_flag']].head(20)
```
